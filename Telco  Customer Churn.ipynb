```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

#Load the customer churn data into a DataFrame

```python
df=pd.read_csv('Customer churn.csv')
df.head()
```

```python
df.info()
```

#replacing blanks with 0 as tenure is 0 and no total charges are recorded

```python
df["TotalCharges"]= df["TotalCharges"].replace(" ","0")
df["TotalCharges"]=df["TotalCharges"].astype("float")
df.info()
```

```python
df.isnull().sum().sum()
```

```python
df.describe()
```

```python
df['customerID'].duplicated().sum()
```

```python
def conv(value):
    if value==1:
        return "Yes"
    else:
      return "No"
df['SeniorCitizen']= df['SeniorCitizen'].apply(conv)
```

#converted 0 or 1 value of senior citizen asto Yes/No to make it easier to understand

```python
df.head()
```

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(x='Churn',data=df)
plt.title("Number of Churned Customer",fontsize=15)
ax.bar_label(ax.containers[0])
plt.show()
```

```python
plt.figure(figsize=(5,5))
gb=df.groupby("Churn").agg({"Churn":"count"})
plt.pie(gb['Churn'],labels=gb.index,autopct="%1.2f%%")
plt.title("Percentage of Churned Customer" ,fontsize=16)
plt.show()
```

#From the given pie chart we can conclude that 26.54% of our Customer have Churned out

```python
plt.figure (figsize=(4,4))
ax=sns.countplot(x="SeniorCitizen",data=df)
ax.bar_label(ax.containers[0])
plt.title("Count of Senior Citizen ",fontsize=14)
plt.show()
```

```python
total_counts=df.groupby("SeniorCitizen")['Churn'].value_counts(normalize=True).unstack()*100
#plot
fig,ax=plt.subplots(figsize=(4,4))   #adjust figsize for better visualization
#plot the bars
total_counts.plot(kind='bar',stacked=True,ax=ax)
#Add Percentage labels on bars
for p in ax.patches:
    width,height=p.get_width(),p.get_height()
    x,y=p.get_xy()
    ax.text(x+width/2,y+height/2, f'{height:.2f}%',ha='center',va='center')

plt.title("Churn by SeniorCitizen(Stacked bar chart)")
plt.xlabel("SeniorCitizen")
plt.ylabel('Percentage(%)')
plt.xticks(rotation=0)
plt.legend(title='Churn',bbox_to_anchor=(0.9,0.9))
plt.show()
```

```python
df.columns.values
```

```python
# Define columns
columns = [ 'PhoneService', 'MultipleLines', 'InternetService',
            'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
            'TechSupport', 'StreamingTV', 'StreamingMovies']

# Number of columns for subplot grid
n_cols = 3
n_rows = (len(columns) + n_cols - 1) // n_cols  # Calculate rows needed

# Create Subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 4))  # Use parentheses for figsize

# Flatten the axes array for easy iteration (handles both 1D & 2D arrays)
axes = axes.flatten()

# Iterate over columns and plot count plots
for i, col in enumerate(columns):
    sns.countplot(x=col, data=df, ax=axes[i], hue=df["Churn"])
    axes[i].set_title(f'Count Plot of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')

# Remove empty subplots (if any)
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
# Customers who do not churn tend to have services like PhoneService, InternetServices (especially DSL), and OnlineSecurity enabled. On the other hand, churn rates are significantly higher when services like OnlineBackup, TechSupport, and StreamingTV are not used or are unavailable, indicating these features are important for customer retention.
```
